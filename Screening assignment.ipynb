```python
#implementation of abstract class in python


from abc import ABC, abstractmethod

class shape(ABC):
    @abstractmethod
    def printarea(self):
        return 0
    
class Rectangle(shape):
    type = "Rectangle"
    sides = 4
    def __init__(self):
        self.length = 6
        self.breadth = 7
    
    def printarea(self):
        return self.length * self.breadth
    
rect1 = Rectangle()
print(rect1.printarea())
```

```python
# implementation of multiple inheritance in python

class Sum:
    def addition(self,a,b):
        return a+b
class Difference:
    def subtraction(self,x,y):
        return x-y
class Product:
    def multiplication(self,m,n):
        return m*n
class derived_Divide(Sum, Difference, Product):
    def division(self,p,q):
        return p/q
    
d = derived_Divide()
print(d.addition(15,21), d.subtraction(23,31), d.multiplication(23,23), d.division(231,33))
```

```python
#implementation of decorator in python
def dosa(func):
    def wrap():
        print ("WELCOME")
        func()
        print ("THANKS FOR LETTING US SERVE YOU")
    return wrap

def ingredients(func):
    def wrap():
        print("CHOOSE FROM THE DOSAS BELOW")
        print ('RAWA DOSA')
        func()
        print ('CHEESE DOSA')
    return wrap

def dosas(food='BUTTER dosa'):
    print(food)

dosas = dosa(ingredients(dosas))
dosas()
```

```python
#implementation of decorator in python

@dosa
@ingredients
def dosas(food='BUTTER DOSA'):
    print(food)

dosas()
```
